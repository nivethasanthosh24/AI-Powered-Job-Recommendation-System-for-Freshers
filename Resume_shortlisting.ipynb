**Load stopwords**

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

**Create text cleaning function**

```python
def clean_text(text):
    # Ensure text is string
    text = str(text)

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    # Split into words
    words = text.split()

    # Remove stopwords
    words = [
        word for word in words
        if word not in stop_words
    ]

    # Join words back
    text = " ".join(words)

    return text
```

**Apply cleaning to dataset**

```python
jobs["clean_description"] = jobs["Job Description"].apply(clean_text)
```

```python
print(jobs[["Job Description", "clean_description"]].head())
```

## **4. Detect Experience Requirement Using NLP (Regex)**

```python
import re
```

**Create Experience Detection Function**

```python
import re
import pandas as pd

def detect_experience(text):

    if pd.isnull(text):
        return 0

    text = str(text).lower()

    # Find all numbers in the text
    numbers = re.findall(r'\d+', text)

    if numbers:
        return int(numbers[0])   # take the first number

    return 0
```

**Apply function to dataset**

```python
jobs["detected_experience"] = jobs["Experience"].apply(detect_experience)
```

**Job type column**

```python
jobs["job_type"] = jobs["detected_experience"].apply(
    lambda x: "Fresher" if x <= 1 else "Experienced"
)
```

## **6. Match User Skills With Jobs**

```python
fresher_jobs = jobs[jobs["job_type"] == "Fresher"]

print("Total Fresher Jobs:", len(fresher_jobs))

fresher_jobs.head()
```

**Create a sample user**

```python
user_skills = "python sql machine learning data analysis"
```

**Convert Skills to Lowercase**

```python
fresher_jobs["skills"] = fresher_jobs["skills"].str.lower()
```

**Create Skill Matching Function**

```python
def skill_match_score(job_skills, user_skills):

    job_words = set(job_skills.split())
    user_words = set(user_skills.split())

    match = job_words.intersection(user_words)

    return len(match)
```

**Apply matching**

```python
fresher_jobs["match_score"] = fresher_jobs["skills"].apply(
    lambda x: skill_match_score(x, user_skills)
)
```

**Get top matching jobs**

```python
recommended_jobs = fresher_jobs.sort_values(
    by="match_score",
    ascending=False
)
```

**Show Top 10 Recommended Jobs**

```python
recommended_jobs[
    ["Job Title", "Company", "skills", "match_score"]
].head(10)
```

## **7. Cosine Similarity**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
```

**Convert Job Skills into Vectors**

```python
job_vectors = vectorizer.fit_transform(
    fresher_jobs["skills"]
)
```

**Convert User Skills into Vector**

```python
user_vector = vectorizer.transform(
    [user_skills]
)
```

**Calculate Similarity**

```python
similarity_scores = cosine_similarity(
    user_vector,
    job_vectors
)
```

**Add Similarity Column**

```python
fresher_jobs["similarity_score"] = similarity_scores.flatten()
```

**Get Top Recommended Jobs**

```python
top_jobs = fresher_jobs.sort_values(
    by="similarity_score",
    ascending=False
)
```

**Show top 10 jobs**

```python
top_jobs[
    ["Job Title", "Company", "skills", "similarity_score"]
].head(10)
```
